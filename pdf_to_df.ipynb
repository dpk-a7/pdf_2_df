```python
pdf_nm = "100100502_AES01267KY_ORG_Financial-Statement_2020.pdf"
```

```python
from tabula.io import read_pdf
df_balance = read_pdf(pdf_nm, pages=4,output_format="dataframe", stream=True)[0]
df_statement = read_pdf(pdf_nm, pages=5,output_format="dataframe", stream=True)[0]
```

```python
df_balance.dropna()
df_balance = df_balance.rename(columns={"As at":"As at 31st March 2020",
                                        "As at.1":"As at 31st March 2019"})
df_balance = df_balance.drop(index=0)
df_balance
```

```python
df_statement = df_statement.rename(columns={"PArtICuLArs":"Particulars", 
                   "Year ended":"Year ended(31st March 2020)",
                   "Year ended.1":"Year ended(31st March 2019)"})
df_statement = df_statement.drop(index=0)
df_statement
```

```python
df_balance.to_csv("Consolidated_Balance.csv")
df_statement.to_csv("Consolidated_statement.csv")
```
